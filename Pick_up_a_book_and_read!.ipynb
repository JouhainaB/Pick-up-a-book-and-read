```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel('/content/Books and HICP pivoted.xlsx')
df
```

```python
df.columns = df.iloc[0]
df=df[1:]
df
```

```python
df.duplicated().sum()
```

```python
df.isna().sum()
```

```python
df.dropna(subset=['Index (2015 = 100)'],inplace=True)
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(data=df,x='Country',y='Index (2015 = 100)')
plt.xticks(rotation=90)
plt.xlabel('Country',fontsize=15)
plt.ylabel('Index',fontsize=15)
plt.title('Country vs Index ',fontsize=16)
plt.grid()
```

```python
sns.histplot(data=df,x='Index (2015 = 100)',y=None,stat='count',bins=30)
plt.xlabel('Index',fontsize=15)
plt.ylabel('count',fontsize=15)
plt.title('Distribution of Index (2015 = 100)',fontsize=16)
plt.show()
```
